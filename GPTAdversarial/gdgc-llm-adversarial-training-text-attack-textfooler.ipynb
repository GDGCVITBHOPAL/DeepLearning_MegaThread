```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification

import torch

tokenizer = AutoTokenizer.from_pretrained("stevhliu/my_awesome_model")

model = AutoModelForSequenceClassification.from_pretrained("stevhliu/my_awesome_model")
model.config.id2label = {0: 'Negative', 1: 'Positive'}

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

model.eval()
```

```python
def predict(text):
    inputs = tokenizer(text, return_tensors="pt")
    inputs = {k: v.to(device) for k, v in inputs.items()}
    
    with torch.no_grad():
        logits = model(**inputs).logits
    
    predicted_class_id = logits.argmax().item()

    return predicted_class_id, model.config.id2label[predicted_class_id]
```

```python
adversarial_examples = [
    # Negative sentiment
    "The movie was not very bad.",
    "I wouldn’t say it was the worst thing ever.",
    "Bad? Maybe. But it had its moments.",
    "Such a masterpiece... said no one ever.",
    "I almost enjoyed it, but not really.",
    "Absolutely amazing—how boring it was.",
    "The acting was on point, if the point was mediocrity.",
    "It was... a movie. That's all I can say.",
    "My dog could’ve directed it better.",
    "It was shockingly ordinary, bordering on sleep-inducing.",
    
    # Positive sentiment
    "Not bad at all.",
    "I thought it was going to be terrible, but I was wrong.",
    "It exceeded my low expectations.",
    "Surprisingly decent, actually.",
    "Okay, I kinda loved it. Don't judge me.",
    "It wasn't Oscar-worthy, but I had a good time.",
    "Honestly, I didn’t hate it.",
    "Started off weak, but finished strong.",
    "So dumb, yet so entertaining.",
    "Guilty pleasure, but a pleasure nonetheless."
]

for i, text in enumerate(adversarial_examples):
    predicted_class_id, predicted_class = predict(text)
    
    print(f"Original Sentiment: {('Positive', 'Negative')[i < 10]}")
    print(f'Text: {text}\nSentiment: {predicted_class}\n')
```

```python
import textattack
import evaluate

import pandas as pd
import numpy as np

from textattack.models.wrappers import HuggingFaceModelWrapper
from textattack.attack_recipes import TextFoolerJin2019
from textattack.datasets import Dataset
from torch.utils.data import DataLoader
from tqdm import tqdm
from transformers import Trainer, TrainingArguments, TextClassificationPipeline, DataCollatorWithPadding
from datasets import Dataset as HFDataset, concatenate_datasets, load_dataset
```

```python
tokenizer = AutoTokenizer.from_pretrained("stevhliu/my_awesome_model")

model = AutoModelForSequenceClassification.from_pretrained("stevhliu/my_awesome_model")
model.config.id2label = {0: 'Negative', 1: 'Positive'}

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

```python
def generate_data(data):
    positive = data.filter(lambda example: example["label"] == 1).select(range(NUM_ADV // 2))
    negative = data.filter(lambda example: example["label"] == 0).select(range(NUM_ADV // 2))
    
    combined = concatenate_datasets([positive, negative])
    combined = combined.shuffle(seed=42)

    return combined
```

```python
dataset = load_dataset("imdb")

train_data = dataset["train"]
test_data = dataset["test"]

combined_train = generate_data(train_data)
combined_test = generate_data(test_data)

train_texts = combined_train["text"]
train_labels = combined_train["label"]

test_texts = combined_test["text"]
test_labels = combined_test["label"]
```

```python
def evaluate_model(model, tokenizer, texts, labels):
    correct = 0
    for text, label in tqdm(zip(texts, labels), total=len(texts)):
        inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(device)
        with torch.no_grad():
            logits = model(**inputs).logits
        pred = logits.argmax(dim=-1).item()
        if pred == label:
            correct += 1
    return correct / len(labels)
```

```python
original_acc = evaluate_model(model, tokenizer, test_texts, test_labels)
print(f"Original accuracy: {original_acc:.4f}")
```

```python
hf_model_wrapper = HuggingFaceModelWrapper(model, tokenizer)
attack = TextFoolerJin2019.build(hf_model_wrapper)

def create_adv_dataset(texts, labels):
    ta_dataset = list(zip(texts, labels))
    ta_dataset = Dataset(ta_dataset)
    
    attack_args = textattack.AttackArgs(num_examples=NUM_ADV, log_to_csv=LOG_PATH, disable_stdout=True)
    attacker = textattack.Attacker(attack, ta_dataset, attack_args)
    attack_results = attacker.attack_dataset()

    advs = []
    for result in attack_results:
        if result.perturbed_result is not None:
            advs.append((result.perturbed_result.attacked_text.text, result.original_result.ground_truth_output))

    return advs
```

```python
advs_train = create_adv_dataset(train_texts, train_labels)
advs_test = create_adv_dataset(test_texts, test_labels)

df_train = pd.DataFrame(advs_train, columns=["text", "label"])
df_test = pd.DataFrame(advs_test, columns=["text", "label"])
```

```python
adv_dataset_train = HFDataset.from_pandas(df_train)
adv_dataset_test = HFDataset.from_pandas(df_test)
```

```python
def tokenize_function(example):
    return tokenizer(example["text"], truncation=True)


def create_model_dataset(raw_dataset):
    adv_dataset = raw_dataset.map(tokenize_function, batched=True)
    adv_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])

    return adv_dataset
```

```python
adv_dataset_train = create_model_dataset(adv_dataset_train)
adv_dataset_test = create_model_dataset(adv_dataset_test)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
accuracy = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    
    return accuracy.compute(predictions=predictions, references=labels)
```

```python
training_args = TrainingArguments(
    output_dir=OUTPUT_PATH,
    per_device_train_batch_size=16,
    num_train_epochs=1,
    eval_strategy="epoch",
    save_strategy="epoch",
    logging_steps=10,
    load_best_model_at_end=True,
    report_to="none"
)

trainer = Trainer(
    model=model, 
    args=training_args, 
    train_dataset=adv_dataset_train,
    eval_dataset=adv_dataset_test,
    processing_class=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

trainer.train()
```

```python
def after_predict(text):
    inputs = adv_tokenizer(text, return_tensors="pt")
    inputs = {k: v.to(device) for k, v in inputs.items()}
    
    with torch.no_grad():
        logits = adv_model(**inputs).logits
    
    predicted_class_id = logits.argmax().item()

    return predicted_class_id, adv_model.config.id2label[predicted_class_id]
```
